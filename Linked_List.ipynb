# Reversing a Linked List

```python
def reverse(l):
  if l.head is None:
    return
  current_node = l.head
  prev_node = None

  while current_node is not None:
  #  Track the next node
    next_node = current_node.next
    # modify the current node
    current_node.next=prev_node

    prev_node=current_node
    current_node=next_node
  l.head = prev_node
```

```python
list2 = LinkedList()
list2.append(2)
list2.append(3)
list2.append(5)
list2.append(9)
```

```python
reverse(list2)
```
